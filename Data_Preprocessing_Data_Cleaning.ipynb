```python
#Checking null values in each column
missing_data=df.isnull().sum()
missing_data
```

```python
#Calculating percentage of data that is null to identify if column should be dropped or not
missing_percentage = (df.isnull().sum() / len(df)) * 100
print("Percentage of missing values per column:")
print(missing_percentage)
```

```python
#Storing and displaying columns that have more than 80% null data
threshold = 80
columns_to_drop = missing_percentage[missing_percentage > threshold].index.tolist()
print(f"\nColumns with more than {threshold}% missing values to be dropped:")
print(columns_to_drop)
```

```python
#Dropping columns with more than 80% null data
df = df.drop(columns=columns_to_drop)
```

```python
#Rechecking information of dataset
df.info()
```

```python
#Rechecking number of rows and columns in dataset
df.shape
```

```python
#Storing numerical and categorical columns seperately
numerical_cols = df.select_dtypes(include=np.number).columns
categorical_cols = df.select_dtypes(include='object').columns
```

```python
#Using sklearn's SimpleImputer filling null values in numerical data with mean and categorical data with most frequent data
from sklearn.impute import SimpleImputer
if len(numerical_cols) > 0:
        numerical_imputer = SimpleImputer(strategy='mean')
        df[numerical_cols] = numerical_imputer.fit_transform(df[numerical_cols])
if len(categorical_cols) > 0:
        categorical_imputer = SimpleImputer(strategy='most_frequent')
        df[categorical_cols] = categorical_imputer.fit_transform(df[categorical_cols])
```

```python
#rechecking for null values in dataset
missing_data = df.isnull().sum()
missing_data
```
