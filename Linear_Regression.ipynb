```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression

#Example dataset(miles_driven and price(k$) )
miles=np.array([100,150,200,250,300]).reshape(-1,1)
price=np.array([50,35,28,15,8])

#Create and train model
model=LinearRegression()
model.fit(miles,price)

#Get slope(coefficient) and intercept
m=model.coef_[0]
c=model.intercept_
print(f"Slope(m): {m}")
print(f"Intercept(c): {c}")

#Predict result values
result_pred=model.predict(miles)

#Plot the data points and regression line
plt.scatter(miles,price,color='blue',label='Actual Price')
plt.plot(miles,result_pred,color='red',label='Expected Price')
plt.xlabel('Miles Driven')
plt.ylabel('Price(k$)')
plt.title('Actual v/s Expected Price of Vehicle')
plt.legend()
plt.show()
```
